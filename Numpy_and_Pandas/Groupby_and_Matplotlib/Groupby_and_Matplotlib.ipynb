```python
import matplotlib.pyplot as plt
%matplotlib inline
A = ['Tom','Bob','John','Monica','Joey']
B = [32,56,76,34,23]
plt.plot(A,B, color = 'Green', linewidth=5, linestyle = 'dashdot')
```

```python
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Age of the people')
plt.plot(A,B, color = 'Orange', linewidth=3, linestyle = 'dotted')
```

```python
X = [1,2,3,4,5,6,7]
Y = [46,56,49,47,47,58,44]
plt.plot(X,Y,'gD-.')
```

```python
plt.plot(X,Y,alpha=0.5) 
#alpha ranges from 0-1, it decides the opaqueness or tranparency.
```

```python
Name = ['Tom','Bob','John','Monica','Joey']
Math_score = [32,56,76,34,23]
Science_score = [78,56,84,93,99]
Finance_score = [53,66,78,65,88]

plt.xlabel ('Names of the students')
plt.ylabel('All subject scores')
plt.title('Overall Grades of Students')

plt.plot(Name,Math_score, label='Math scores')
plt.plot(Name,Science_score, label='Science scores')
plt.plot(Name,Finance_score, label='finance scores')

plt.legend(loc = 'best', shadow=True, fontsize='small')

plt.grid()
```

## Bar Plot Matplotlib

```python
company = ['Google','Amazon','Microsoft','Meta']
revenue = [90,128,98,88]
profit = [40,2,34,12]
```

```python
import numpy as np
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.ylabel('revenue(bln)')
plt.title('US Tech stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.legend()
```

```python
plt.ylabel('revenue(bln)')
plt.title('US Tech stocks')
plt.bar(company,revenue,label='Revenue')
plt.legend()
```

```python
plt.ylabel('revenue(bln)')
plt.title('US Tech stocks')              # difference
plt.bar(company,revenue,label='Revenue')
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel('revenue(bln)')
plt.title('US Tech stocks')

plt.bar(ypos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(ypos+0.2, profit, width=0.4, label='Profit')

plt.legend()
```

```python
plt.yticks(ypos,company)
plt.title('US Tech stocks')

plt.barh(ypos-0.2, revenue, label='Revenue')
plt.barh(ypos+0.2, profit, label='Profit')

plt.legend()
```

## Histogram Matplotlib

```python
blood_sugar = [183,85,90,150,149,99,88,98,115,135,77,82,129]

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth = 0.95, color = 'g')
```

```python
men_blood_sugar = [183,85,90,150,149,99,88,98,115,135,77,82,129]
women_blood_sugar = [65,83,65,79,90,115,121,128,86,96,88,93,87]

plt.xlabel("Sugar range")
plt.ylabel('Total no. of patients')
plt.title('Suger range of patients')

plt.hist([men_blood_sugar, women_blood_sugar], bins=[80,100,125,150], rwidth = 0.95, 
         color = ['green','orange'], label = ['Men','Women'])

plt.legend()
```

## Pie Chart Matplotlib

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home rent','Food','Phone/internet bill','Car','Other utilities']
plt.axis('equal')    # Otherwise there is a pixel change and it comes as oval shape
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct='%0.1f%%', shadow=True,
       explode=[0,0.2,0,0,0], startangle=45)
plt.show()
```
